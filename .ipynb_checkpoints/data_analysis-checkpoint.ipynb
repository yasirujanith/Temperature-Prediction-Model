```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

%matplotlib inline
from matplotlib.pylab import rcParams

rcParams['figure.figsize'] = 15, 5

import datetime

from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error

import statsmodels.api as sm

from statsmodels.tsa.stattools import adfuller
```

```python
# plt.figure(figsize=(10,10))
```

```python
df = pd.read_csv('india_temperatures.csv')
df.head()
```

```python
df.shape
```

```python
new_df = pd.DataFrame(columns=['month', 'temp'])
```

```python
for i in range(df.shape[0]):
    for j in range(1, 13):
        new_df.loc[i*12 + j] = [str(int(df.loc[i]['YEAR'])) + '-' + "{0:0=2d}".format(j), df.iloc[i,j]]
new_df.head()
```

```python
new_df['month'] = pd.to_datetime(new_df['month'])
new_df.head()
```

```python
new_df.set_index('month', inplace=True)
new_df.head()
```

```python
new_df.shape
```

```python
ts = new_df['temp']
```

```python
plt.figure(figsize=(15,5))
plt.plot(ts)
```

```python
plt.figure(figsize=(15,5))
plt.plot(ts.loc['2000-01-01':])
```

```python
plt.figure(figsize=(7,7))
pd.plotting.lag_plot(ts)
```

```python
decomposition = sm.tsa.seasonal_decompose(ts, model='additive')
decomposition.plot()
plt.show()
```

```python
ts = ts.loc['1950-01-01':]
```

```python
ts_lg = pd.concat([ts, ts.shift(1), ts.shift(2)], axis=1)
ts_lg.columns = ['t', 't-1', 't-2']
result = ts_lg.corr()
print('Correlation between t & t =', result['t']['t'])
print('Correlation between t & t-1 =', result['t']['t-1'])
print('Correlation between t & t-2 =', result['t']['t-2'])
```

```python
split = ts.index.min() + (ts.index.max() - ts.index.min())/2

ts_1, ts_2 = ts[ts.index.min():split], ts[split:]
mean1, mean2 = ts_1.mean(), ts_2.mean()
var1, var2 = ts_1.var(), ts_2.var()
print('mean1=%f, mean2=%f' % (mean1, mean2))
print('variance1=%f, variance2=%f' % (var1, var2))
```

```python
# p-value > 0.05 = non-stationary
# p-value <= 0.05 = stationary

result = adfuller(ts)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```

```python
dif_ts = ts - ts.shift(1)
dif_ts = dif_ts - dif_ts.shift(1)
dif_ts.dropna(inplace=True)

plt.figure(figsize=(15,5))
plt.plot(dif_ts.loc['2000-01-01':])
```

```python
result = adfuller(dif_ts)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```

```python
train, test = ts[0:len(ts)-10], ts[len(ts)-10:]

print(ts.shape)
print(train.shape)
print(test.shape)
```

```python
history = [x for x in train]
predictions = test.copy(deep=True)

for t in range(len(test)):
    model = ARIMA(history, order=(12,1,2))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions[t] = yhat
    obs = test[t]
    history.append(yhat)
    print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
```

```python
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```
